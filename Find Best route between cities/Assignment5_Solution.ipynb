```python
# Reading from city_data.csv with pandas
import pandas as pd
city_distances = pd.read_csv('city_data.csv', header=None)
city_distances
```

```python
#Read data with pandas with column and index name
no_of_cities = city_distances.shape[0]
no_of_cities
```

```python
city_distances.info()
```

```python
# Calculating the min distance using floyd warshall algo
def calculate_min_distances(city_distances):
    intermediate_city_distances = city_distances
    for intermediate_city in range(no_of_cities):
        for city1 in range(no_of_cities):
            for city2 in range(no_of_cities):
                if intermediate_city_distances[intermediate_city][city1] + intermediate_city_distances[intermediate_city][city2] < intermediate_city_distances[city1][city2]:
                    intermediate_city_distances[city1][city2] = intermediate_city_distances[intermediate_city][city1] + intermediate_city_distances[intermediate_city][city2]
    return intermediate_city_distances
min_city_distances = calculate_min_distances(city_distances)
min_city_distances
```

```python
def print_all_distances(shortest_distances, city_names):
    for city1 in range(no_of_cities):
        for city2 in range(city1, no_of_cities):
            print(f"The shortest distance between {city_names[city1]} and {city_names[city2]} is {shortest_distances[city1][city2]}")
```
